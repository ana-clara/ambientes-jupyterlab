```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
with open('/content/metal_bands_2017.csv', encoding="utf8", errors="ignore") as f:
    df = pd.read_csv(f)
```

```python
df = df.dropna()
```

```python
df.columns = np.array(["id", "band_name", "fans", "formed", "origin", "split", "style"])
sorted_df = df.sort_values(by='origin')
```

```python
countries = []
number_per_country = []

for index, row in sorted_df.iterrows():
    if row['origin'] in countries:
        number_per_country[countries.index(row['origin'])] += 1
    else:
        countries.append(row['origin'])
        number_per_country.append(1)
```

```python
plt.figure(figsize=[25,10])
plt.bar(countries, number_per_country, width=1.2, color="magenta")
plt.xticks(fontsize=5, rotation=45)
plt.show()
```
